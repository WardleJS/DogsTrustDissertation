```python
import pandas as pd

data = pd.read_csv('Master_Modelling.csv')

# Function to create a standardized color combination
def combine_colours(row):
    if pd.isna(row['Colour_2']) or row['Colour_2'] == '':
        # If Colour_2 is missing, use only Colour_main
        return row['Colour_main']
    else:
        # Otherwise, combine Colour_main and Colour_2, sorted alphabetically
        return ' & '.join(sorted([row['Colour_main'], row['Colour_2']]))

# Apply the function to each row
data['Colour_Combination'] = data.apply(combine_colours, axis=1)

# Count the occurrences of each colour combination
colour_counts = data['Colour_Combination'].value_counts()

colour_counts_df = colour_counts.reset_index()
colour_counts_df.columns = ['Colour_Combination', 'Count']

colour_counts_df.to_csv('colour_combinations_counts.csv', index=False)

print("CSV file 'colour_combinations_counts.csv' has been saved.")
```
